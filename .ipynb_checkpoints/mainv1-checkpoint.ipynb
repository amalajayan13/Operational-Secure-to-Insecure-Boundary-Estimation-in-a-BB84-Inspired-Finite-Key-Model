# Quantum Cryptographic Security as a Phase Transition: Q1 Upgraded Variational Notebook

```python

# ============================================================
# 1. Environment setup, reproducibility, and output folders
# Title cell only, followed by Python cells as requested.
# This upgraded notebook extends the starter branch with:
#   - dense control-parameter sweeps
#   - critical-point detection
#   - susceptibility analysis
#   - scaling-law estimation
#   - scenario trajectories and figure/table export
# ============================================================

import os, json, math, warnings, itertools, textwrap
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")
np.random.seed(42)

IN_COLAB = False
try:
    import google.colab  # type: ignore
    from google.colab import drive  # type: ignore
    IN_COLAB = True
except Exception:
    IN_COLAB = False

if IN_COLAB:
    DRIVE_ROOT = "/content/drive"
    MOUNT_POINT = f"{DRIVE_ROOT}/MyDrive"
    try:
        if not os.path.exists(MOUNT_POINT):
            drive.mount(DRIVE_ROOT, force_remount=True)
        else:
            drive.mount(DRIVE_ROOT, force_remount=False)
    except Exception as e:
        print(f"[WARN] Drive mount issue: {e}")
    PROJECT_ROOT = "/content/drive/MyDrive/Outputs/QSFE_phase_transition_q1"
else:
    PROJECT_ROOT = "/mnt/data/QSFE_phase_transition_q1"

FIG_DIR = os.path.join(PROJECT_ROOT, "figures")
TAB_DIR = os.path.join(PROJECT_ROOT, "tables")
OTH_DIR = os.path.join(PROJECT_ROOT, "others")

for d in [PROJECT_ROOT, FIG_DIR, TAB_DIR, OTH_DIR]:
    os.makedirs(d, exist_ok=True)

print("[PATH CHECK] PROJECT_ROOT =", PROJECT_ROOT)
print("[PATH CHECK] FIG_DIR exists:", os.path.exists(FIG_DIR))
print("[PATH CHECK] TAB_DIR exists:", os.path.exists(TAB_DIR))
```

```python

# ============================================================
# 2. Scientific configuration
# The control parameter eta plays the role of an effective
# stress/noise variable driving a secure-to-insecure transition.
# ============================================================

TIME_HORIZON = 120
T = np.arange(TIME_HORIZON, dtype=float)

EPS = 1e-9
QBER_CRIT = 0.110
LAMBDA_ENTROPY = 1.35
MU_INSTABILITY = 0.90
NU_RESOURCE = 0.35

# Dense control sweep for phase transition analysis.
ETA_GRID = np.linspace(0.01, 0.55, 90)
ATTACK_LEVELS = [0.15, 0.30, 0.50]
DRIFT_BASE = 0.015

SCENARIOS = [
    {"name": "stable_low_noise",     "eta": 0.06, "attack": 0.12, "drift": 0.02, "curvature": 0.0000},
    {"name": "moderate_transition",  "eta": 0.13, "attack": 0.25, "drift": 0.03, "curvature": 0.0004},
    {"name": "critical_approach",    "eta": 0.18, "attack": 0.38, "drift": 0.04, "curvature": 0.0009},
    {"name": "late_collapse",        "eta": 0.16, "attack": 0.28, "drift": 0.04, "curvature": 0.0016},
]

BB84_STAGES = [
    "State preparation",
    "Quantum transmission",
    "Measurement and basis reconciliation",
    "Parameter estimation",
    "Error correction",
    "Privacy amplification"
]

BB84_INTERPRETATION = {
    "State preparation": "Initial calibration, source quality, and state fidelity.",
    "Quantum transmission": "Environmental disturbance and channel-induced decoherence.",
    "Measurement and basis reconciliation": "Observed mismatch and effective QBER realization.",
    "Parameter estimation": "Entropy and exposure inference under finite observations.",
    "Error correction": "Leakage overhead, correction burden, and residual inconsistencies.",
    "Privacy amplification": "Final secrecy extraction and operational robustness margin."
}

print("[CONFIG] ETA range:", (float(ETA_GRID.min()), float(ETA_GRID.max())))
print("[CONFIG] Time horizon:", TIME_HORIZON)
print("[BB84] stages:", " | ".join(BB84_STAGES))
```

```python

# ============================================================
# 3. Core theory functions
# These functions instantiate the variational / phase-transition
# view of quantum security.
# ============================================================

def clip01(x):
    return np.clip(x, 0.0, 1.0)

def binary_entropy(p):
    p = np.clip(p, EPS, 1 - EPS)
    return -(p * np.log2(p) + (1 - p) * np.log2(1 - p))

def qber_from_eta(eta, attack, drift, time_index):
    seasonal = 0.012 * np.sin(2 * np.pi * time_index / max(TIME_HORIZON, 1))
    trend = 0.0015 * time_index / max(TIME_HORIZON - 1, 1)
    qber = eta + 0.030 * attack + 0.40 * drift + seasonal + trend
    return clip01(qber)

def entropy_resource(qber, drift):
    # Resource decreases with QBER and implementation drift.
    return clip01(1.0 - binary_entropy(np.minimum(qber, 0.499)) - 0.85 * drift)

def adversarial_exposure(attack, qber, drift):
    raw = 0.62 * attack + 0.24 * (qber / max(QBER_CRIT, EPS)) + 0.14 * (drift / 0.20)
    return clip01(raw)

def instability_term(qber_series, attack_series, k):
    if k <= 1:
        return 0.0
    dq = qber_series[k] - qber_series[k-1]
    da = attack_series[k] - attack_series[k-1]
    return float(np.sqrt(dq**2 + 0.60 * da**2))

def order_parameter(S, E):
    denom = np.maximum(S + E, EPS)
    return (S - E) / denom

def free_energy(S, E, temperature):
    return E - temperature * S

def classify_phase(psi):
    if psi > 0.10:
        return "secure"
    if psi > -0.10:
        return "critical"
    return "insecure"

def susceptibility(psi_series, control_series):
    return np.gradient(psi_series, control_series)

def detect_first_critical_time(psi_series):
    idx = np.where(psi_series <= 0.0)[0]
    return int(idx[0]) if len(idx) else -1

def action_integral(L_series, delta_t=1.0):
    return float(np.sum(L_series[:-1] * delta_t))
```

```python

# ============================================================
# 4. Trajectory simulation engine
# Generates full dynamical trajectories for one effective
# control parameter eta.
# ============================================================

def simulate_trajectory(eta, attack, drift, curvature=0.0, time_horizon=TIME_HORIZON):
    tt = np.arange(time_horizon, dtype=float)
    qber = np.zeros(time_horizon)
    attack_series = np.zeros(time_horizon)
    drift_series = np.zeros(time_horizon)

    for k in range(time_horizon):
        attack_series[k] = clip01(attack + 0.10 * np.sin(2*np.pi*k/max(time_horizon,1)) + 0.002*k)
        drift_series[k] = clip01(drift + curvature * (k**2) / max((time_horizon-1)**2, 1))
        qber[k] = qber_from_eta(eta, attack_series[k], drift_series[k], k)

    S = np.array([entropy_resource(qber[k], drift_series[k]) for k in range(time_horizon)])
    E = np.array([adversarial_exposure(attack_series[k], qber[k], drift_series[k]) for k in range(time_horizon)])
    I = np.array([instability_term(qber, attack_series, k) for k in range(time_horizon)])
    temperature = 1.0 + 1.60 * qber + 0.20 * attack_series
    F = free_energy(S, E, temperature)
    Psi = order_parameter(S, E)
    L = E - LAMBDA_ENTROPY * S + MU_INSTABILITY * I + NU_RESOURCE * qber
    Phi = -np.gradient(F, tt)
    Gamma = np.gradient(np.gradient(F, tt), tt)
    Chi = np.gradient(Psi, tt)

    return pd.DataFrame({
        "t": tt,
        "eta": eta,
        "attack": attack_series,
        "drift": drift_series,
        "qber": qber,
        "S": S,
        "E": E,
        "I": I,
        "temperature": temperature,
        "F": F,
        "L": L,
        "Psi": Psi,
        "Phi": Phi,
        "Gamma": Gamma,
        "Chi_t": Chi,
        "phase": [classify_phase(x) for x in Psi]
    })

# Quick smoke test.
smoke_df = simulate_trajectory(eta=0.10, attack=0.20, drift=0.03, curvature=0.0002)
print(smoke_df.head(3))
print("[SMOKE] final Psi =", round(float(smoke_df['Psi'].iloc[-1]), 4))
```

```python

# ============================================================
# 5. Global phase landscape over (eta, attack)
# This is the macroscopic phase portrait of the framework.
# ============================================================

landscape_rows = []
for eta, attack in itertools.product(ETA_GRID, ATTACK_LEVELS):
    drift = DRIFT_BASE + 0.07 * eta
    traj = simulate_trajectory(eta=float(eta), attack=float(attack), drift=float(drift), curvature=0.0003)
    final = traj.iloc[-1]
    landscape_rows.append({
        "eta": float(eta),
        "attack_level": float(attack),
        "final_qber": float(final["qber"]),
        "mean_S": float(traj["S"].mean()),
        "mean_E": float(traj["E"].mean()),
        "mean_F": float(traj["F"].mean()),
        "final_Psi": float(final["Psi"]),
        "min_Psi": float(traj["Psi"].min()),
        "max_abs_Gamma": float(np.abs(traj["Gamma"]).max()),
        "first_critical_time": detect_first_critical_time(traj["Psi"].values),
        "phase": classify_phase(float(final["Psi"]))
    })

landscape_df = pd.DataFrame(landscape_rows)
landscape_csv = os.path.join(TAB_DIR, "phase_landscape_q1.csv")
landscape_df.to_csv(landscape_csv, index=False)

plt.figure(figsize=(8, 5))
for attack in ATTACK_LEVELS:
    sub = landscape_df[landscape_df["attack_level"] == attack].sort_values("eta")
    plt.plot(sub["eta"], sub["final_Psi"], label=f"attack={attack:.2f}")
plt.axhline(0.0, linestyle="--")
plt.xlabel("Control parameter $\\eta$")
plt.ylabel("Final order parameter $\\Psi$")
plt.title("Security phase diagram: final order parameter vs control parameter")
plt.legend()
phase_diagram_path = os.path.join(FIG_DIR, "critical_phase_diagram.png")
plt.tight_layout()
plt.savefig(phase_diagram_path, dpi=220)
plt.close()

print("[LANDSCAPE] rows =", len(landscape_df))
print("[LANDSCAPE] saved to:", landscape_csv)
print("[FIGURE] saved:", phase_diagram_path)
```

```python

# ============================================================
# 6. Critical-point detection and susceptibility analysis
# The critical point eta_c is estimated where final Psi crosses 0.
# ============================================================

critical_rows = []

for attack in ATTACK_LEVELS:
    sub = landscape_df[landscape_df["attack_level"] == attack].sort_values("eta").copy()
    eta_vals = sub["eta"].values
    psi_vals = sub["final_Psi"].values
    chi_vals = np.gradient(psi_vals, eta_vals)
    sub["Chi_eta"] = chi_vals

    # Linear interpolation for Psi = 0 crossing, if available.
    eta_c = np.nan
    for i in range(len(psi_vals) - 1):
        if psi_vals[i] == 0:
            eta_c = eta_vals[i]
            break
        if psi_vals[i] * psi_vals[i+1] < 0:
            x0, x1 = eta_vals[i], eta_vals[i+1]
            y0, y1 = psi_vals[i], psi_vals[i+1]
            eta_c = x0 - y0 * (x1 - x0) / (y1 - y0)
            break

    critical_rows.append({
        "attack_level": float(attack),
        "eta_c": float(eta_c) if not np.isnan(eta_c) else np.nan,
        "max_abs_chi": float(np.max(np.abs(chi_vals))),
        "eta_at_max_abs_chi": float(eta_vals[int(np.argmax(np.abs(chi_vals)))]),
    })

    sub.to_csv(os.path.join(TAB_DIR, f"critical_sweep_attack_{attack:.2f}.csv"), index=False)

critical_df = pd.DataFrame(critical_rows)
critical_csv = os.path.join(TAB_DIR, "critical_points_summary.csv")
critical_df.to_csv(critical_csv, index=False)

plt.figure(figsize=(8, 5))
for attack in ATTACK_LEVELS:
    sub = landscape_df[landscape_df["attack_level"] == attack].sort_values("eta")
    chi_vals = np.gradient(sub["final_Psi"].values, sub["eta"].values)
    plt.plot(sub["eta"], np.abs(chi_vals), label=f"attack={attack:.2f}")
plt.xlabel("Control parameter $\\eta$")
plt.ylabel(r"$|\chi| = \left|\partial \Psi / \partial \eta\right|$")
plt.title("Security susceptibility vs control parameter")
plt.legend()
chi_path = os.path.join(FIG_DIR, "susceptibility_divergence.png")
plt.tight_layout()
plt.savefig(chi_path, dpi=220)
plt.close()

print(critical_df)
print("[FIGURE] susceptibility:", chi_path)
```

```python

# ============================================================
# 7. Scaling-law analysis near the critical point
# We estimate an exponent beta from |Psi| ~ |eta - eta_c|^beta
# using the first attack branch that exhibits a crossing.
# ============================================================

scaling_records = []
selected_attack = None
selected_eta_c = None

for _, row in critical_df.iterrows():
    if not np.isnan(row["eta_c"]):
        selected_attack = float(row["attack_level"])
        selected_eta_c = float(row["eta_c"])
        break

if selected_attack is None:
    print("[SCALING] No critical crossing found; scaling analysis skipped.")
else:
    sub = landscape_df[landscape_df["attack_level"] == selected_attack].sort_values("eta").copy()
    eps = np.abs(sub["eta"].values - selected_eta_c)
    psi_abs = np.abs(sub["final_Psi"].values)

    mask = (eps > 0.004) & (eps < 0.10) & (psi_abs > 1e-3)
    x = eps[mask]
    y = psi_abs[mask]

    if len(x) >= 5:
        coeff = np.polyfit(np.log(x), np.log(y), 1)
        beta = float(coeff[0])
        intercept = float(coeff[1])

        for xx, yy in zip(x, y):
            scaling_records.append({
                "attack_level": selected_attack,
                "eta_c": selected_eta_c,
                "distance_to_critical": float(xx),
                "abs_final_Psi": float(yy)
            })

        scaling_df = pd.DataFrame(scaling_records)
        scaling_csv = os.path.join(TAB_DIR, "scaling_law_points.csv")
        scaling_df.to_csv(scaling_csv, index=False)

        plt.figure(figsize=(6, 5))
        plt.loglog(x, y, marker="o", linestyle="", label="data")
        fit_y = np.exp(intercept) * x**beta
        plt.loglog(x, fit_y, linestyle="-", label=fr"fit: $\beta={beta:.3f}$")
        plt.xlabel(r"$|\eta - \eta_c|$")
        plt.ylabel(r"$|\Psi|$")
        plt.title("Scaling law near the critical point")
        plt.legend()
        scaling_fig = os.path.join(FIG_DIR, "scaling_law_loglog.png")
        plt.tight_layout()
        plt.savefig(scaling_fig, dpi=220)
        plt.close()

        scaling_meta = {
            "selected_attack": selected_attack,
            "eta_c": selected_eta_c,
            "beta": beta,
            "intercept": intercept,
            "n_points": int(len(x))
        }
        with open(os.path.join(OTH_DIR, "scaling_metadata.json"), "w", encoding="utf-8") as f:
            json.dump(scaling_meta, f, indent=2)

        print("[SCALING] selected attack =", selected_attack)
        print("[SCALING] eta_c =", round(selected_eta_c, 5))
        print("[SCALING] beta =", round(beta, 5))
    else:
        print("[SCALING] Not enough near-critical points for a stable fit.")
```

```python

# ============================================================
# 8. Representative trajectory evaluation
# These scenario runs provide narrative-ready figures and tables.
# ============================================================

trajectory_frames = []
summary_rows = []

for sc in SCENARIOS:
    df = simulate_trajectory(
        eta=sc["eta"],
        attack=sc["attack"],
        drift=sc["drift"],
        curvature=sc["curvature"]
    ).copy()
    df["scenario"] = sc["name"]
    trajectory_frames.append(df)

    summary_rows.append({
        "scenario": sc["name"],
        "final_qber": float(df["qber"].iloc[-1]),
        "mean_entropy_resource": float(df["S"].mean()),
        "mean_exposure": float(df["E"].mean()),
        "mean_free_energy": float(df["F"].mean()),
        "final_action": float(action_integral(df["L"].values)),
        "min_Psi": float(df["Psi"].min()),
        "max_abs_Gamma": float(np.abs(df["Gamma"]).max()),
        "max_susceptibility_time": int(np.argmax(np.abs(df["Chi_t"]).values)),
        "first_critical_time": detect_first_critical_time(df["Psi"].values)
    })

traj_df = pd.concat(trajectory_frames, ignore_index=True)
traj_csv = os.path.join(TAB_DIR, "trajectory_evaluation_q1.csv")
traj_df.to_csv(traj_csv, index=False)

summary_df = pd.DataFrame(summary_rows)
summary_csv = os.path.join(TAB_DIR, "trajectory_summary_q1.csv")
summary_df.to_csv(summary_csv, index=False)

print(summary_df.to_string(index=False))
print("[TABLE] saved:", summary_csv)
```

```python

# ============================================================
# 9. Save scenario figures
# Each scenario gets: Psi(t), free-energy indicators, and a
# trajectory phase portrait in (qber, Psi, F).
# ============================================================

for scenario_name, sub in traj_df.groupby("scenario"):
    sub = sub.sort_values("t")

    plt.figure(figsize=(7, 4.5))
    plt.plot(sub["t"], sub["Psi"])
    plt.axhline(0.0, linestyle="--")
    plt.xlabel("Time")
    plt.ylabel(r"Order parameter $\Psi(t)$")
    plt.title(f"Order parameter trajectory: {scenario_name}")
    path = os.path.join(FIG_DIR, f"psi_{scenario_name}.png")
    plt.tight_layout()
    plt.savefig(path, dpi=220)
    plt.close()

    plt.figure(figsize=(7, 4.5))
    plt.plot(sub["t"], sub["F"], label="F(t)")
    plt.plot(sub["t"], sub["Phi"], label="Phi(t)")
    plt.plot(sub["t"], sub["Gamma"], label="Gamma(t)")
    plt.xlabel("Time")
    plt.ylabel("Indicator value")
    plt.title(f"Free-energy indicators: {scenario_name}")
    plt.legend()
    path = os.path.join(FIG_DIR, f"free_energy_indicators_{scenario_name}.png")
    plt.tight_layout()
    plt.savefig(path, dpi=220)
    plt.close()

    plt.figure(figsize=(6, 5))
    plt.scatter(sub["qber"], sub["Psi"], c=sub["F"], s=25)
    plt.axhline(0.0, linestyle="--")
    plt.xlabel("QBER")
    plt.ylabel(r"$\Psi$")
    plt.title(f"Phase portrait colored by free energy: {scenario_name}")
    path = os.path.join(FIG_DIR, f"phase_portrait_{scenario_name}.png")
    plt.tight_layout()
    plt.savefig(path, dpi=220)
    plt.close()

print("[FIGURES] scenario figures exported to", FIG_DIR)
```

```python

# ============================================================
# 10. Global aggregate visualizations
# ============================================================

pivot_df = landscape_df.pivot(index="eta", columns="attack_level", values="final_Psi")

plt.figure(figsize=(7, 5))
for attack in ATTACK_LEVELS:
    sub = landscape_df[landscape_df["attack_level"] == attack].sort_values("eta")
    plt.scatter(sub["mean_E"], sub["mean_S"], label=f"attack={attack:.2f}", s=15)
plt.xlabel("Mean exposure")
plt.ylabel("Mean entropy resource")
plt.title("Macroscopic security landscape in (E, S) space")
plt.legend()
global_portrait_path = os.path.join(FIG_DIR, "global_security_landscape_ES.png")
plt.tight_layout()
plt.savefig(global_portrait_path, dpi=220)
plt.close()

plt.figure(figsize=(8, 5))
for attack in ATTACK_LEVELS:
    sub = landscape_df[landscape_df["attack_level"] == attack].sort_values("eta")
    plt.plot(sub["eta"], sub["mean_F"], label=f"attack={attack:.2f}")
plt.xlabel("Control parameter $\\eta$")
plt.ylabel("Mean free energy")
plt.title("Mean free energy across the control sweep")
plt.legend()
global_F_path = os.path.join(FIG_DIR, "mean_free_energy_vs_eta.png")
plt.tight_layout()
plt.savefig(global_F_path, dpi=220)
plt.close()

print("[FIGURE] saved:", global_portrait_path)
print("[FIGURE] saved:", global_F_path)
```
